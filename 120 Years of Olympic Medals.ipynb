```python
clean_olympic=olympic.drop(['Height','Weight','Games', 'City', 'Sex'], axis=1)
clean_olympic
```

```python
clean_olympic.set_index('ID', inplace=True)
```

```python
clean_olympic['Team_as_Num'] = clean_olympic['Team_as_Num'].replace(np.nan, 0)
```

```python
mapping_dict = {
    'NaN':0,
    'Bronze':1,
    'Silver':2,
    'Gold':3
}

clean_olympic['Medal_as_Num']=clean_olympic['Medal'].map(mapping_dict)
clean_olympic['Medal_as_Num'].replace(np.nan, 0)
```

```python
# To find the average age of Olympians
clean_olympic['Age'].dropna().mean()
```

```python
clean_olympic.fillna({ 'Age': 26, 'Medal': 'None', 'Medal_as_Num': 0}, inplace=True)
```

```python
# Changing all of the floats into integers
columns_to_convert = ['Team_as_Num', 'Medal_as_Num', 'Age']
clean_olympic[columns_to_convert] = clean_olympic[columns_to_convert].astype(int)
```

```python
clean_olympic
```

```python
# Checking to make sure all NaN/nulls are gone
clean_olympic.isnull().sum()
```

```python
top_50_winter_olympic=clean_olympic[(clean_olympic['Season'] == 'Winter') & (clean_olympic['Team_as_Num'] != 0)] #see disclaimer
top_50_winter_olympic
```

```python
team_medal_count_by_winter_sport = top_50_winter_olympic[['Team_as_Num','Team','Sport','Medal', 'Medal_as_Num']]\
                                  .groupby(['Team_as_Num', 'Team', 'Sport', 'Medal']) \
                                  .count()
team_medal_count_by_winter_sport.head(20)
```

```python
top_50_summer_olympic=clean_olympic[(clean_olympic['Season'] == 'Summer') & (clean_olympic['Team_as_Num'] != 0)] #see disclaimer
top_50_summer_olympic
```

```python
team_medal_count_by_summer_sport = top_50_summer_olympic[['Team_as_Num','Team','Sport', 'Medal', 'Medal_as_Num']]\
                                  .groupby(['Team_as_Num', 'Team', 'Sport', 'Medal']) \
                                  .count()
team_medal_count_by_summer_sport.tail(20)
```

```python
alpine_ski=top_50_winter_olympic[(top_50_winter_olympic['Sport'] == 'Alpine Skiing') & (top_50_winter_olympic['Medal_as_Num']!= 0)] #see disclaimer
diving=top_50_summer_olympic[(top_50_summer_olympic['Sport'] == 'Diving') & (top_50_summer_olympic['Medal_as_Num']!=0)]#see disclaimer

diving
```

```python
diving_medals= diving[['Team_as_Num','Team', 'Medal', 'Medal_as_Num']]\
                                  .groupby(['Team_as_Num', 'Team', 'Medal']) \
                                  .count()
diving_medals
```

```python
ice_hockey=top_50_winter_olympic[(top_50_winter_olympic['Sport'] == 'Ice Hockey') & (top_50_winter_olympic['Medal_as_Num']!= 0)] #see disclaimer
ice_hockey.drop(['Season', 'Event'], axis=1)
```
